```python
# Check if an integer is an instance of the int class
print(isinstance(5, int))  # Output: True
print(type(5)==int)

# Check if a string is an instance of the str class
print(isinstance("hello", str))  # Output: True

# Check if a float is an instance of the int class
print(isinstance(3.14, int))  # Output: False
```

```python
# Check if an object is an instance of any type in a tuple of types
print(isinstance(5, (int, float)))  # Output: True (5 is an int, which is in the tuple)

print(isinstance("hello", (int, float, str)))  # Output: True (string is in the tuple)
```
