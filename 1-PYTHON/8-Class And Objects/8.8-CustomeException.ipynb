```python
class Error(Exception):
    pass
class DobException(Error):
    pass
```

```python
year = int(input("Enter The dob : "))
age = 2024 - year

try :
    if age>20 and age<30 :
        print("Valid Age ")
    else :
        raise DobException
except :
    print("Your Age Doesn't Belong to the teenager category")
```
