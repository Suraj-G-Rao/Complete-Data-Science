```python
from transformers import pipeline
classifier = pipeline("sentiment-analysis")
```

```python
def func(utterance):
    return classifier(utterance)
```

```python
import gradio as gr
descriptions = "this is sentiment analysis app"
```

```python
app = gr.Interface(fn=func,inputs="text",outputs="text",title="sentiment analysis",description=descriptions)  
app.launch(share=True)
```
